Experiments and Feature Showcases

```python
import FNTVer1
import Graph
import secrets
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
```

Draw a FNT graph

```python
rate = 6
size = 100
fnt = FNTVer1.FishingNet(rate)
for i in range(size):
    fnt.nextNode(secrets.token_hex(16), datetime.now())

Graph.drawFNT(fnt)
```

Show the cumulative weight of all nodes

```python
Graph.drawCWs(fnt)
```

Show the growth of CW of a node as number of nodes increases

```python
Graph.drawCWChg(fnt, 10)
```

Show the throughput of each layer

```python
Graph.drawThroughput(fnt)
```

Show the utilization of each layer

```python
Graph.drawUtilization(fnt)
```

Show the waste rate of each layer

```python
Graph.drawWasteRate(fnt)
```

Show the cumulative waste rate

```python
Graph.drawCumWaste(fnt)
```

For the same Rate, in the case of nodes of different orders of magnitude, the throughput waste comparison

```python
f1 = FNTVer1.FishingNet(5)
f1.count = 100
f2 = FNTVer1.FishingNet(5)
f2.count = 200
f3 = FNTVer1.FishingNet(5)
f3.count = 500
f4 = FNTVer1.FishingNet(5)
f4.count = 1000

cws1 = f1.findCumWaste()
cws2 = f2.findCumWaste()
cws3 = f3.findCumWaste()
cws4 = f4.findCumWaste()

l1 = np.arange(1, len(cws1) + 1)
l2 = np.arange(1, len(cws2) + 1)
l3 = np.arange(1, len(cws3) + 1)
l4 = np.arange(1, len(cws4) + 1)

plt.figure(1, figsize=(10, 6))

plt.rc('font', size=20)
plt.rc('axes', titlesize=20)
plt.rc('axes', labelsize=20)
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
plt.rc('legend', fontsize=20)
plt.rc('figure', titlesize=20)

plt.title("Comparison of Waste Rate Under Different Number of Nodes")
plt.xlabel("Layers")
plt.ylabel("Cumulative Waste Rate")
plt.ylim(0, 1.1)

plt.plot(l1, cws1, color='red', label='r=5, S=100')
plt.text(l1[-1], cws1[-1], "{:.2%}".format(cws1[-1]), ha='center', va='bottom')
plt.plot(l2, cws2, color='orange', label='r=5, S=200')
plt.text(l2[-1], cws2[-1], "{:.2%}".format(cws2[-1]), ha='center', va='top')
plt.plot(l3, cws3, color='green', label='r=5, S=500')
plt.text(l3[-1], cws3[-1], "{:.2%}".format(cws3[-1]), ha='center', va='bottom')
plt.plot(l4, cws4, color='blue', label='r=5, S=1000')
plt.text(l4[-1], cws4[-1], "{:.2%}".format(cws4[-1]), ha='center', va='bottom')

plt.legend()
plt.show()

f1 = FNTVer1.FishingNet(10)
f1.count = 100
f2 = FNTVer1.FishingNet(10)
f2.count = 1000
f3 = FNTVer1.FishingNet(10)
f3.count = 10000
f4 = FNTVer1.FishingNet(10)
f4.count = 100000

cws1 = f1.findCumWaste()
cws2 = f2.findCumWaste()
cws3 = f3.findCumWaste()
cws4 = f4.findCumWaste()

l1 = np.arange(1, len(cws1) + 1)
l2 = np.arange(1, len(cws2) + 1)
l3 = np.arange(1, len(cws3) + 1)
l4 = np.arange(1, len(cws4) + 1)

plt.figure(1, figsize=(10, 6))

plt.rc('font', size=20)
plt.rc('axes', titlesize=20)
plt.rc('axes', labelsize=20)
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
plt.rc('legend', fontsize=20)
plt.rc('figure', titlesize=20)

plt.title("Comparison of Waste Rate Under Different Number of Nodes")
plt.xlabel("Layers")
plt.ylabel("Cumulative Waste Rate")
plt.ylim(0, 1.1)

plt.plot(l1, cws1, color='red', label='r=10, S=100')
plt.text(l1[-1], cws1[-1], "{:.2%}".format(cws1[-1]), ha='center', va='top')
plt.plot(l2, cws2, color='orange', label='r=10, S=1000')
plt.text(l2[-1], cws2[-1], "{:.2%}".format(cws2[-1]), ha='center', va='bottom')
plt.plot(l3, cws3, color='green', label='r=10, S=10,000')
plt.text(l3[-1], cws3[-1], "{:.2%}".format(cws3[-1]), ha='center', va='top')
plt.plot(l4, cws4, color='blue', label='r=10, S=100,000')
plt.text(l4[-1], cws4[-1], "{:.2%}".format(cws4[-1]), ha='center', va='bottom')

plt.legend()
plt.show()
```

Conclusion: In the case of the same rate, the larger the number of nodes, the smaller the waste

Conclusion: For nodes of different orders of magnitude, there is a rate that is most suitable for it to maximize the utilization of the network (minimize the waste)
