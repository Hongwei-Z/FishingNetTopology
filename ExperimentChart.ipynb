Experiments and Feature Showcases

```python
import FNT
import Graph
import secrets
from datetime import datetime
```

Define rate and number of nodes

```python
rate = 10
size = 100
```

Create a FishingNet, add nodes

```python
fnt = FNT.FishingNet(rate)
for i in range(size):
    fnt.nextNode(secrets.token_hex(16), datetime.now())
```

Draw a FNT graph

```python
Graph.drawFNT(fnt)
```

Show the cumulative weight of all nodes

```python
Graph.drawCWs(fnt)
```

Show the growth of CW of a node as number of nodes increases

```python
Graph.drawCWChg(fnt, 10)
```

Show the throughput of each layer

```python
Graph.drawThroughput(fnt)
```

Show the utilization of each layer

```python
Graph.drawUtilization(fnt)
```

Show the waste rate of each layer

```python
Graph.drawWasteRate(fnt)
```
