Experiments and Feature Showcases

```python
import FNT
import Graph
import secrets
from datetime import datetime
```

Draw a FNT graph

```python
rate1 = 10
size1 = 1000
f1 = FNT.FishingNet(rate1)
for i in range(size1):
    f1.nextNode(secrets.token_hex(16), datetime.now())

rate2 = 6
size2 = 100
f2 = FNT.FishingNet(rate2)
for j in range(size2):
    f2.nextNode(secrets.token_hex(16), datetime.now())

Graph.drawFNT(f1)
Graph.drawFNT(f2)
```

Show the cumulative weight of all nodes

```python
Graph.drawCWs(f2)
```

Show the growth of CW of a node as number of nodes increases

```python
Graph.drawCWChg(f1, 800)
Graph.drawCWChg(f2, 10)
```

Show the throughput of each layer

```python
Graph.drawThroughput(f1)
Graph.drawThroughput(f2)
```

Show the utilization of each layer

```python
Graph.drawUtilization(f1)
Graph.drawUtilization(f2)
```

Show the waste rate of each layer

```python
Graph.drawWasteRate(f1)
Graph.drawWasteRate(f2)
```

Show the cumulative waste rate

```python
Graph.drawCumWaste(f1)
Graph.drawCumWaste(f2)
```
