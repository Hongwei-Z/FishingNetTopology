Experiments and Feature Showcases

```python
import FNT
import Visualization
import secrets
from datetime import datetime
```

Define rate and number of nodes

```python
rate = 6
nodes = 100
```

Create a FishingNet, add nodes

```python
fnt = FNT.FishingNet(rate)
for i in range(nodes):
    fnt.nextNode(secrets.token_hex(16), datetime.now())
fnt.toString()
```

```python
Visualization.draw_FNT(rate, nodes)
```

Find a node and show the result

```python
print("Node 50:", fnt.findNode(50))
fnt.findNode(50).toString()
```

Find tips for a node and show the result

```python
print("Two tips:", fnt.findTipsIndex(50))
print("Two tips:", fnt.findTipsData(50))
```

Find the two nodes that approve this node

```python
print(fnt.findApprove(80))
```

Find the cumulative weight of a node

```python
fnt.findCW(80)
```

Disable a node

```python
fnt.disableNode(80)
fnt.toString()
```
