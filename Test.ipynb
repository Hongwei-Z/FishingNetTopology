Experiments and Feature Showcases

```python
import FNT
import Graph
import secrets
from datetime import datetime
```

Define rate and number of nodes

```python
rate = 6
size = 100
```

Create a FishingNet, add nodes

```python
fnt = FNT.FishingNet(rate)
for i in range(size):
    fnt.nextNode(secrets.token_hex(16), datetime.now())
fnt.printFNT()
```

```python
Graph.drawFNT(rate, size)
```

Find a node and show the result

```python
fnt.findNode(80)
```

Find tips for a node and show the result

```python
fnt.findTips(80)
```

Find the two nodes that approve this node

```python
fnt.findApprove(80)
```

Find the cumulative weight of a node

```python
fnt.findCW(80)
```

Print all nodes that have directly or indirectly approved this node

```python
fnt.findSubnet(80)
```

Draw a line chart to show cumulative weight for each node

```python
Graph.drawCWs(fnt)
```

Draw the graph to show the change of the CW of a node as the # of nodes increases

```python
Graph.drawCWChg(rate, size, 15)
```

Disable a node, only disconnect from other nodes, not delete data

```python
fnt.disableNode(80)
fnt.printFNT()
```
